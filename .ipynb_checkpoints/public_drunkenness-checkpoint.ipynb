## Dataframe with urination and public drunkenness complaints

Here's where we get the initial data.

### Filtering out unwanted data

For this analysis I removed complaints of underage drinking since it relates less to my particular topic. I also only kept complaints made between 8 pm and 4 am, which I decided constitutes a timeframe for nighttime activity. Note that it's possible that my filtering for public urination and drinking doesn't include all relevant complaints since there are some 311 complaints that are catch-alls, like "Emergency Response Team (ERT)" and "Miscellaneous Categories."

```python
# Filtering out unwanted data
df.drop(df[df.descriptor == 'After Hours - Licensed Est'].index, inplace=True)
df.drop(df[df.descriptor == 'Underage - Licensed Est'].index, inplace=True)
```

```python
df['created_date'] = pd.to_datetime(df['created_date'])
df['created_time'] = df['created_date'].dt.time
```

```python
df = df[(df['created_time'] >= time(20,00,00)) | (df['created_time'] <= time(4,00,00))]
```

### Spatial transformation

Here's where we create spatial geometries that will be used later on to map the data.

```python
df['latitude'] = df['latitude'].astype(float)
df['longitude'] = df['longitude'].astype(float)
```

```python
df['lonlat'] = list(zip(df.longitude, df.latitude))
```

```python
df['geometry'] = df[['lonlat']].applymap(lambda x:shapely.geometry.Point(x))
```

```python
gdf = gpd.GeoDataFrame(df)
gdf.crs = from_epsg(4326)
gdf.head(5)
```

## Yelp API query

Midway through the project, I decided to add in the locations of dive bars in Manhattan to look for patterns between bar locations and 311 complaints. To get the data, I used Yelp's API to collect information on every one dollar sign "dive bar" located in Manhattan. To execute the API call, I relied on code provided here: https://github.com/Yelp/yelp-fusion/blob/master/fusion/python/sample.py.

Note that you'll need a Yelp API key stored in a text file called "yelp_api.txt" to run the following code.

```python
# You no longer need to provide Client ID to fetch Data
# It now uses private keys to authenticate requests (API Key)
# You can find it on
# https://www.yelp.com/developers/v3/manage_app
API_KEY = open('yelp_api.txt', 'r').readlines()[0][:] # api stored in seperate .txt file

# API constants, you shouldn't have to change these.
API_HOST = 'https://api.yelp.com'
SEARCH_PATH = '/v3/businesses/search'
BUSINESS_PATH = '/v3/businesses/'  # Business ID will come after slash.

# Defaults for our simple example.
DEFAULT_TERM = 'dinner'
DEFAULT_LOCATION = 'San Francisco, CA'
SEARCH_LIMIT = 50

def request(host, path, api_key, url_params=None):
    """Given your API_KEY, send a GET request to the API.
    Args:
        host (str): The domain host of the API.
        path (str): The path of the API after the domain.
        API_KEY (str): Your API Key.
        url_params (dict): An optional set of query parameters in the request.
    Returns:
        dict: The JSON response from the request.
    Raises:
        HTTPError: An error occurs from the HTTP request.
    """
    url_params = url_params or {}
    url = '{0}{1}'.format(host, quote(path.encode('utf8')))
    headers = {
        'Authorization': 'Bearer %s' % api_key,
    }

    response = requests.request('GET', url, headers=headers, params=url_params)

    return response.json()

def search(api_key, term, location, price, offset):
    """Query the Search API by a search term and location.
    Args:
        term (str): The search term passed to the API.
        location (str): The search location passed to the API.
    Returns:
        dict: The JSON response from the request.
    """

    url_params = {
        'term': term.replace(' ', '+'),
        'location': location.replace(' ', '+'),
        'limit': SEARCH_LIMIT,
        'price': price,
        'offset': offset
    }
    return request(API_HOST, SEARCH_PATH, api_key, url_params=url_params)

def get_business(api_key, business_id):
    """Query the Business API by a business ID.
    Args:
        business_id (str): The ID of the business to query.
    Returns:
        dict: The JSON response from the request.
    """
    business_path = BUSINESS_PATH + business_id

    return request(API_HOST, business_path, api_key)

def query_api(term, location, price, offset):
    """Queries the API by the input values from the user.
    Args:
        term (str): The search term to query.
        location (str): The location of the business to query.
    """
    response = search(API_KEY, term, location, price, offset)

    businesses = response.get('businesses')
    return businesses
```

```python
# Create offsets for multiple API calls
offsets = [0]
for i in range(1,20):
    off = i*50
    offsets.append(off)
```

```python
# Loop through multiple queries to get all one-dollar dive bars in Manhattan
results = []
for i in range(len(offsets)):
    dives = query_api('dive bar', 'Manhattan, NY', '1', offsets[i])
    results.append(dives)
```

```python
# Check to make sure all dive bars have been collected
# Since we see that after the 8th call there are no results, we can assume that we have all the dive bars
for i in range(len(results)):
    print(len(results[i]))
```

```python
# Make lists of the bar names, lats, and longs since I don't need all the info in the dictionary
names = []
lats = []
lons = []
for i in range(len(results)):
    for j in range(len(results[i])):
        name = results[i][j]['name']
        lat = results[i][j]['coordinates']['latitude']
        lon = results[i][j]['coordinates']['longitude']
        names.append(name)
        lats.append(lat)
        lons.append(lon)
```

```python
# Make a dataframe
dive_df = pd.DataFrame(
    {'name': names,
     'lat': lats,
     'lon': lons
    })
```

### Regression

I also exported the data from ArcGIS to run a series of linear regression models looking for patterns or correlations between the number of 311 complaints and the number of dive bars per neighborhood.

```python
final_data = pd.read_csv("final_data.csv")
```

```python
final_data.head(5)
```

```python
# Sample of neighborhood names
print(final_data.ntaname)
```

```python
# Initial regression plotting the number of complaints (normalized by neighborhood area) against the number of dive bars
# Note that the number of complaints and dive bars were scaled up by a factor of 100,000 since the figures were very small and 
# therefore difficult to interpret.
ols = smf.ols(formula='comps_area_x100k ~ dives_area_x100k', data=final_data).fit()
ols.summary()
```

```python
pl.figure(figsize=(10,8))
pl.scatter(final_data['dives_area_x100k'],final_data['comps_area_x100k'], c='r')
pl.plot(final_data['dives_area_x100k'], ols.predict(final_data),'-', color='blue')
pl.xlabel('Dive Bars (10e5)', fontsize=15)
pl.ylabel('311 Complaints (10e5)', fontsize=15)
pl.title('311 Complaints Against Dive Bars per Neighborhood', fontsize=15);
```

```python
# Dataframe that only includes neighborhoods with at least one dive bar
final_data_mod = final_data[final_data['dives_area_x100k'] > 0]
```

```python
final_data_mod.head(5)
```

```python
ols2 = smf.ols(formula='comps_area_x100k ~ dives_area_x100k', data=final_data_mod).fit()
ols2.summary()
```

```python
pl.figure(figsize=(10, 8))
pl.plot(final_data_mod['dives_area_x100k'],final_data_mod['comps_area_x100k'],'go')
pl.plot(final_data_mod['dives_area_x100k'], ols2.predict(final_data_mod),'-', color='orange')
pl.xlabel('Dive Bars (10e5)', fontsize=15)
pl.ylabel('311 Complaints (10e5)', fontsize=15)
pl.title('311 Complaints Against Dive Bars per Neighborhood', fontsize=15);
```

```python
pl.figure(figsize=(10, 8))
pl.scatter(final_data['dives_area_x100k'],final_data['comps_area_x100k'], c='r', alpha=.2, label='all dives')
pl.scatter(final_data_mod['dives_area_x100k'],final_data_mod['comps_area_x100k'],c='g', alpha=.2, label='dives>0')
pl.plot(final_data['dives_area_x100k'], ols.predict(final_data),'-', color='blue', label='corr1')
pl.plot(final_data_mod['dives_area_x100k'], ols2.predict(final_data_mod),'-', color='orange', label='corr2')
pl.xlabel('Dive Bars (10e5)', fontsize=15)
pl.ylabel('311 Complaints (10e5)', fontsize=15)
pl.legend()
pl.title('311 Complaints Against Dive Bars per Neighborhood', fontsize=15);
```

## Conclusion

Based on the distribution of public drunkenness, public urination, and dive bars in Manhattan, we can see that New Yorkers who go out to drink at night prefer to go to Lower Manhattan and north of Central Park. We also saw that there is a positive correlation between the number of 311 complaints relating to public drunkenness and public urination depending on the number of cheap dive bars in a neighborhood (as long as that neighborhood has at least one cheap dive bar). Of course, there are many other factors that explain an increase in 311 complaints for a particular neighborhood. But for now, we can at least say with some certainty that dive bars have a meaningful impact on the amount of public drunkenness that takes place in a given neighborhood.
